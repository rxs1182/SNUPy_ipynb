## Imports

```python
# Data Manipulation imports
import pandas as pd
import numpy as np

# Data Analysis imports
import statistics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Graphing imports
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.io as pio
pio.renderers.default = "browser"
```

## Data Preprocessing

```python
# Reading the csv into a pandas dataframe
df = pd.read_csv("/Users/rohansingh/Downloads/cs labs - sorting.csv")
```

```python
df
```

```python
# Obtaining the feature variable
x = df["array size"].dropna().to_numpy().reshape(-1,1)
```

```python
# Obtaining the dependent variable
y = df[pd.to_numeric(df["insertion sort"], errors='coerce').notnull()]["insertion sort"].to_numpy().reshape(-1,1)
```

## Graphical Analysis

```python
# Getting the dataframe of the plotting data
df_data = pd.DataFrame(list(zip(df["array size"].dropna().to_list(),df[pd.to_numeric(df["insertion sort"], errors='coerce').notnull()]["insertion sort"].to_list())), columns=["size","time"])
df_data
```

```python
# Plotting it out using plotly express
fig = px.line(df_data,x="size",y="time",title="Insertion Sort Analysis")
fig.show()
```

## Regression Analysis

```python
# Test_Train split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
# Fitting it into a linear regression model

lin_reg_model = LinearRegression()

lin_reg_model.fit(x_train,y_train)

lin_reg_model.score(x_test, y_test)
```

```python
# Fitting it into a polynomial regression model

poly = PolynomialFeatures(degree=2, include_bias=False)

# Changing x to x^2
poly_features = poly.fit_transform(x_train)
poly_test = poly.fit_transform(x_test)

poly_reg_model = LinearRegression()

poly_reg_model.fit(poly_features, y_train)

poly_reg_model.score(poly_test,y_test)
```
