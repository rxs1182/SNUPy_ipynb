## Task 2

### 1. Frequency of polls by each pollster

#### Approval Poll Frequency Stats

```python
# Obtaining the list of unique pollsters
d1_pollster_list = list(set(df_1["pollster"].to_list()))

print("There are ",len(d1_pollster_list), " pollsters for Dataset 1")

# Initializing the frequency list
frequency_list = []
for i in range(0,len(d1_pollster_list),1):
    frequency_list.append(0)
    
pollster_list = df_1["pollster"].to_list()
for i in range(0,len(pollster_list),1):
    # Classifying the pollster
    frequency_list[d1_pollster_list.index(pollster_list[i])] += 1

# Appending the values into a table
df_d1_frequency = pd.DataFrame(list(zip(d1_pollster_list,frequency_list)), columns=["Pollster","Frequency"])
df_d1_frequency
```

#### Concern Poll Frequency

```python
# Obtaining the list of unique pollsters
d2_pollster_list = list(set(df_2["pollster"].to_list()))

print("There are ",len(d2_pollster_list), " pollsters for Dataset 2")

# Initializing the frequency list
frequency_list = []
for i in range(0,len(d2_pollster_list),1):
    frequency_list.append(0)
    
pollster_list = df_2["pollster"].to_list()
for i in range(0,len(pollster_list),1):
    # Classifying the pollster
    frequency_list[d2_pollster_list.index(pollster_list[i])] += 1

# Appending the values into a table
df_d2_frequency = pd.DataFrame(list(zip(d2_pollster_list,frequency_list)), columns=["Pollster","Frequency"])
df_d2_frequency
```

### 2. Aggregated proportions of respondents approving of Donald Trump’s performance

```python
# Grouping Donald Trumps performance by party
# Assuming that "all" and "I" are different

parties = list(set(df_1["party"].to_list()))

# Counts for the total people approving
r_approval = 0
i_approval = 0
all_approval = 0
d_approval = 0

# Sample sizes of the polls
r_sample_size = 0
i_sample_size = 0
all_sample_size = 0
d_sample_size = 0

# Iterating through all of the entries
sample_sizes = df_1["sample_size"].to_list()
approval_ratings = df_1["approve"].to_list()
party = df_1["party"].to_list()

for i in range(0,len(sample_sizes),1):
    # Classifying by party
    if(party[i] == "R"):
        r_approval += approval_ratings[i]*sample_sizes[i]
        r_sample_size += sample_sizes[i]
    elif(party[i] == "I"):
        i_approval += approval_ratings[i]*sample_sizes[i]
        i_sample_size += sample_sizes[i]
    elif(party[i] == "all"):
        all_approval += approval_ratings[i]*sample_sizes[i]
        all_sample_size += sample_sizes[i]
    else:
        d_approval += approval_ratings[i]*sample_sizes[i]
        d_sample_size += sample_sizes[i]
    


# Appending aggregated ratings
aggregated_ratings = []
aggregated_ratings.append(r_approval/r_sample_size)
aggregated_ratings.append(i_approval/i_sample_size)
aggregated_ratings.append(all_approval/all_sample_size)
aggregated_ratings.append(d_approval/d_sample_size)

# Appending Sample Sizes
aggregated_samples = []
aggregated_samples.append(r_sample_size)
aggregated_samples.append(i_sample_size)
aggregated_samples.append(all_sample_size)
aggregated_samples.append(d_sample_size)

# Making a datframe for our findings
df_approval_rating = pd.DataFrame(list(zip(parties,aggregated_ratings, aggregated_samples)),columns=["Party","Aggregated Approval Rating","Aggregate Sample Size"])
df_approval_rating
```

### 3. Aggregate worries about the economy

```python
# ['concern-infected', 'concern-economy']

# Variables to store the count of the worries and sample size
very_worried_count_1 = 0
somewhat_worried_count_1 = 0
sample_size_count_1 = 0

# Iterating through all of the entries
very_worried_list = df_2["very"].to_list()
somewhat_worried_list = df_2["somewhat"].to_list()
sample_size_list = df_2["sample_size"].to_list()
subject_list = df_2["subject"].to_list()

for i in range(0,len(sample_size_list),1):
    # Checking if the Subject of the is economy
    if subject_list[i] == "concern-economy":
        # Adding the counts
        very_worried_count_1 += very_worried_list[i]*sample_size_list[i]
        somewhat_worried_count_1 += somewhat_worried_list[i]*sample_size_list[i]
        sample_size_count_1 += sample_size_list[i]

# Organizing results
aggregates = []
aggregates.append(very_worried_count_1/sample_size_count_1)
aggregates.append(somewhat_worried_count_1/sample_size_count_1)

titles =["very worried","somewhat worried"]

# Making a dataframe
df_economy = pd.DataFrame(list(zip(titles,aggregates)),columns=["Concern_Level","Aggregate Propoportions"])
df_economy
```
